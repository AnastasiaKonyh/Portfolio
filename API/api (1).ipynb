```python
import requests 
import pandas as pd
import json
```

```python
url = 'https://api.vk.com/method/wall.get'
params = {
        'domain': 'skyeng',
        'access_token': token,
        'v':'5.131',
        'count': 100,
        'filter': 'all'
      
    }
    
req = requests.get(url, params=params)
res = req.json()

print('–ö–æ–ª-–≤–æ –∑–∞–ø–∏—Å–µ–π:', len(pd.DataFrame(res['response'])))
```



–í—ã–≥—Ä—É–∂–∞–µ–º –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é `posts` –≤—Å–µ –∑–∞–ø–∏—Å–∏ —Å–æ–æ–±—â–µ—Å—Ç–≤–∞  



```python
import time
offset = 0
max_count = 100
posts = pd.DataFrame()
while True:
    url = 'https://api.vk.com/method/wall.get'
    params = {
        'domain': 'skyeng',
        'access_token': token,
        'v':'5.131',
        'count': max_count,
        'offset': offset
      
    }
    req = requests.get(url, params=params)
    res = req.json()
    if len (res['response']['items'])!=0:
        posts = pd.concat([posts, pd.DataFrame(res['response']['items'])])
    else:
        break
    print('–£–∂–µ —Å–∫–∞—á–∞–Ω–æ', offset, '–∑–∞–ø–∏—Å–µ–π')
    offset = offset + max_count
    time.sleep(0.5)
        
```

```python
print('–ö–æ–ª-–≤–æ –∑–∞–ø–∏—Å–µ–π:', len(posts))
```



–í—ã–≥—Ä—É–∂–∞–µ–º 50 —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–æ—Å—Ç–æ–≤. –ú–µ—Ç—Ä–∏–∫–æ–π –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏ –º—ã —Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É –ª–∞–π–∫–æ–≤, —Ä–µ–ø–æ—Å—Ç–æ–≤ –∏ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤.


```python
posts['comments_count'] = posts['comments'].apply(lambda x: x['count'])
posts['likes_count'] = posts['likes'].apply(lambda x: x['count'])
posts['reposts_count'] = posts['reposts'].apply(lambda x: x['count'])
posts.head()
```

```python
# –∑–∞–ø–æ–ª–Ω—è–µ–º –¥–∞–Ω–Ω—ã–º–∏ —Å—Ç–æ–ª–±–µ—Ü —Å —Å—É–º–º–æ–π –ø–æ –≤—Å–µ–º –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—è–º

posts['activity'] =posts['comments_count']+ posts['likes_count']+posts['reposts_count']
posts.head()
```

```python
posts_50 = posts.sort_values(by='activity', ascending = False).head(50)
posts_50.head()
```



–ü–æ—á–∏—Å—Ç–∏–º –¥–∞–Ω–Ω—ã–µ, —Å–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
