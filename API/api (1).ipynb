Теперь перед тобой стоит задача – выгрузить маркетологам для анализа 50 самых популярных постов за все время существования сообщества. Метрику популярности будем считать как сумму лайков, репостов и комментариев к каждому посту. Если останется время, для дополнительного анализа выгрузи все комментарии к этим постам в отдельный файл.

```python
# читаем токен из файла my_own_token.py, куда мы сохранили access_token
from my_token import my_token as token
```



__Сформируем запрос, который получает 100 записей со стены сообщества__



```python
import requests 
import pandas as pd
import json
```

```python
url = 'https://api.vk.com/method/wall.get'
params = {
        'domain': 'skyeng',
        'access_token': token,
        'v':'5.131',
        'count': 100,
        'filter': 'all'
      
    }
    
req = requests.get(url, params=params)
res = req.json()

print('Кол-во записей:', len(pd.DataFrame(res['response'])))
```



Выгружаем в переменную `posts` все записи сообщества  



```python
import time
offset = 0
max_count = 100
posts = pd.DataFrame()
while True:
    url = 'https://api.vk.com/method/wall.get'
    params = {
        'domain': 'skyeng',
        'access_token': token,
        'v':'5.131',
        'count': max_count,
        'offset': offset
      
    }
    req = requests.get(url, params=params)
    res = req.json()
    if len (res['response']['items'])!=0:
        posts = pd.concat([posts, pd.DataFrame(res['response']['items'])])
    else:
        break
    print('Уже скачано', offset, 'записей')
    offset = offset + max_count
    time.sleep(0.5)
        
```

```python
print('Кол-во записей:', len(posts))
```

```python
posts.head()
```



Выгружаем 50 самых популярных постов. Метрикой популярности мы считаем сумму лайков, репостов и комментариев.


```python
posts['comments_count'] = posts['comments'].apply(lambda x: x['count'])
posts['likes_count'] = posts['likes'].apply(lambda x: x['count'])
posts['reposts_count'] = posts['reposts'].apply(lambda x: x['count'])
posts.head()
```

```python
# заполняем данными столбец с суммой по всем активностям

posts['activity'] =posts['comments_count']+ posts['likes_count']+posts['reposts_count']
posts.head()
```

```python
posts_50 = posts.sort_values(by='activity', ascending = False).head(50)
posts_50.head()
```



Почистим данные, сохраняем результат




```python
posts_50 = posts_50[:50]
postsTable = posts_50[['id', 'date', 'text', 'comments_count', 'likes_count', 'reposts_count', 'activity']]
postsTable.head()
```
